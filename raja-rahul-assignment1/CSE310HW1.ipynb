```python
import matplotlib.pyplot as plt
import numpy as np

## all times are in ms (recorded for a trial of 10 times)
facebook1 = [137,135,131,121,126,123,130,133,153,117]
facebook2 = [9,10,11,4,9,10,2,9,9,10]
facebook3 = [6,6,5,13,13,13,25,6,13,13]

tiktok1 = [248,234,233,236,239,239,239,248,256,219]
tiktok2 = [22,9,11,9,9,30,10,10,9,10]
tiktok3 = [38,25,17,14,22,25,30,15,15,17]

amazon1 = [107,110,109,107,105,107,109,102,101,102]
amazon2 = [9,4,9,10,11,3,10,2,10,4]
amazon3 = [40,31,41,24,31,27,32,22,22,21]

data = [[facebook1, tiktok1, amazon1],[facebook2, tiktok2, amazon2],[facebook3, tiktok3, amazon3]]

fig = plt.figure(figsize = (10, 7))
ax = fig.add_axes([0,0,1,1])


x = ["", "Facebook", "Tiktok", "Amazon"]
default_x_ticks = range(len(x))
plt.xticks(default_x_ticks, x)
c="purple"
bp1 = plt.boxplot(data[0], notch=True, patch_artist=True,
            boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )


c = "red"
bp2 = plt.boxplot(data[1],  notch=True, patch_artist=True,
            boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )
c = "blue"
bp3 = plt.boxplot(data[2],notch=True, patch_artist=True,
            boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )
ax.legend([bp1["boxes"][0], bp2["boxes"][0], bp3["boxes"][0]], ["mydig", "dig", "Google DNS Resolver"], loc='upper right')
plt.show()
```
